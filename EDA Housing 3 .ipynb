# Outlier Detection

```python
plt.boxplot(df["SalePrice"])
plt.show()
```

```python
df["SalePrice"].describe()
```

```python
print(df['SalePrice'].quantile(0.10))
print(df['SalePrice'].quantile(0.90))
```
